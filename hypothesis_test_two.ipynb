## Hypothesis Test

สมมติว่าชุดข้อมุลดังกล่าวเป็น Normal distribution
พิสูจน์ว่าค่าใช้จ่ายเฉลี่ยในเดือนมกราคมกับเดือนกุมภาพันธ์เท่ากันหรือไม่ โดย 
- H0: ค่าใช้จ่ายเฉลี่ยในเดือนมกราคมกับเดือนกุมภาพันธ์เท่ากัน
- Ha: ค่าใช้จ่ายเฉลี่ยในเดือนมกราคมกับเดือนกุมภาพันธ์ไม่เท่ากัน

จาก Test statistic: t = (x - y − Δ0) / sqrt(s1^2/m + s2^2/n)

```python
t_s = (mean_jan - mean_feb - 0) / math.sqrt((sd_jan**2)/len(df_jan)+(sd_feb**2)/len(df_feb))
t_s
```

degree of freedom


```python
v = (((sd_jan**2)/len(df_jan))+((sd_feb**2)/len(df_feb)))**2 / ((((sd_jan**2 / len(df_jan))**2)/(len(df_jan)-1))+(((sd_feb**2/len(df_feb))**2)/(len(df_feb)-1)))
v
```

```python
len(df_feb)
```

```python
from scipy.stats import t
t.ppf(0.01/2,v)
```

```python
t.cdf(t_s,v)
```

เนื่องจาก p-value = 0.056 > alpha = 0.01 ----> not reject null

ดังนั้นสรุปได้ว่า ค่าใช้จ่ายเฉลี่ยในเดือนมกราคมกับเดือนกุมภาพันธ์เท่ากัน
